```python
# Create table query
create_table_query1 = """
CREATE TABLE IF NOT EXISTS length_of_stay_constraint (
    Patient_ID INTEGER PRIMARY KEY,
    Age INTEGER CHECK (Age >= 0),
    Gender TEXT CHECK (Gender IN ('Male', 'Female', 'Other')),
    Admission_Type TEXT NOT NULL,
    Primary_Diagnosis TEXT,
    Comorbidities TEXT,
    Previous_Hospitalizations INTEGER CHECK (Previous_Hospitalizations >= 0),
    BP TEXT,
    Heart_rate INTEGER CHECK (Heart_rate BETWEEN 40 AND 200),
    Surgical_Procedures TEXT,
    Severity_of_Illness TEXT CHECK (Severity_of_Illness IN ('Mild', 'Moderate', 'Severe', 'Critical')),
    Insurance_Type TEXT NOT NULL,
    Socioeconomic_Status TEXT CHECK (Socioeconomic_Status IN ('Upper class', 'Below poverty line',
    'Lower class', 'Middle class')),
    Hospital_Total_Beds INTEGER CHECK (Hospital_Total_Beds > 0),
    Room_Type TEXT CHECK (Room_Type IN ('Ward', 'Semiprivate', 'Private')),
    Discharge_Destination TEXT CHECK (Discharge_Destination IN ('Rehabilitation facility',
    'Deceased', 'Longterm care facility',
       'Home with family')),
    Length_of_Stay_in_days INTEGER CHECK (Length_of_Stay_in_days >= 0)
);
"""
```

```python
#  query to execute create a database with try catch
try:

  conn = psycopg2.connect(DATABASE_URL)
  cursor = conn.cursor()

  cursor.execute(create_table_query1)

  conn.commit()
  cursor.close()
  conn.close()

except Exception as e:
  print(e)
```
